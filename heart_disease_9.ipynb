```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
x = data.drop(columns= 'target', axis=1)
y = data['target']
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.20,random_state=2,stratify=y)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
x_train_prediction = model.predict(x_train)
```

```python
accuracy_score(x_train_prediction, y_train)
```

```python

#training data accuracy
training_data_acc = accuracy_score(x_train_prediction,y_train)
print('Training Data Accuracy:' , training_data_acc)
x_test_prediction = model.predict(x_test)
test_data_acc = accuracy_score(x_test_prediction, y_test)
print ('Test Data Accuracy:', test_data_acc)
```
